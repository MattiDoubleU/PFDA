 1. Import data frames, load the data set & analyse rows and columns:

```python
# Data frames.
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data set.
file_path = r"C:\Users\User\Desktop\ATU\PFDA\PFDA-courseware\assignment\weatherreadings1.csv"
df = pd.read_csv(file_path)

# Ran df without below but it wouldn't sow the entire data. This is the solution:
# https://stackoverflow.com/questions/49188960/how-to-show-all-columns-names-on-a-large-pandas-dataframe

pd.options.display.max_columns = None
pd.options.display.max_rows = 5 # Wanted to show 5 rows only.

# Let's have a look.

df
```

 2. As per assignment instructions row *'dryBulbTemperature_Celsius'* identified and plotted over *'createdDateTime'* the most useful datapoint (row) for this exercise:

```python
# Convert 'createdDateTime' column to datetime.
df['createdDateTime'] = pd.to_datetime(df['createdDateTime'], errors='coerce')

# Plot the temperature over time.
plt.figure(figsize=(12, 7))
plt.plot(df['createdDateTime'], df['dryBulbTemperature_Celsius'], marker='o', linestyle='-', color='#6b2e21')

# Format the x-axis to display hours and minutes. https://matplotlib.org/stable/api/_as_gen/matplotlib.axis.Axis.set_major_formatter.html.
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))

# Auto format the dates to prevent overlap.
plt.gcf().autofmt_xdate()

# Add titles and labels.
plt.title('Temperature Over Time', fontsize=18)
plt.xlabel('Date & Time', fontsize=14)
plt.ylabel('Temperature (°C)', fontsize=14)

# Format the x-axis for better readability.
plt.xticks(rotation=40)

# Show grid lines.
plt.grid(True)

# Show the plot.
plt.tight_layout()
plt.show()
```
